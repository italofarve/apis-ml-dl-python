# Test 6. Usando palabras claves. Primera prueba

```
import requests
import json
from requests.auth import HTTPBasicAuth

if __name__ == '__main__':
  url = 'https://www.udemy.com/api-2.0/courses/'

  client_id = 'Coloca_aquí_tu_credencial'
  client_secret = 'Coloca_aquí_tu_credencial'
  authentication = HTTPBasicAuth(client_id, client_secret)

  # En payload almacenamos las palabras claves a filtrar. 
  payload = {'search': 'aprende python', 'ratings': '4'}
  response = requests.get(url, auth=authentication, params=payload)

  # Estado 403 indica error (no tenemos permiso). Estado 200 indica llamada correcta
  print("Estado de respuesta:", response.status_code)
  print(response.url)
  
  # Si estado es 200 leemos el resultado
  if response.status_code == 200:
    response_json = json.loads(response.text)
    count = response_json['count']
    print('numero de resultados arrojados:', count)
    print('response_json')
    print("_____Resultados____")
    print(json.dumps(response_json, indent=2))
    
    # Guardamos los resultados
    content = response.content
    file = open('test6_bd.json', 'wb')
    file.write(content)
    file.close()
```

```
# Creamos un diccionario. Será nuestro payload

claves = ['search', 'ratings']
valores = []

print('Para hacer una consulta necesitamos algunos parametros:')
print('search: Filter courses by a search term. Igual que haces en Google')
print('ratings: Filter courses by minimum average rating', "\n")
print('A continuación de te pediremos {} parametros'.format(len(claves)))

for x in range(len(claves)):
    #valores.append( input("Introduce un valor: ") )
    valores.append(input('Introduce {} '.format(claves[x])))
print(valores)

# Resultado tiene que ser: payload = {'search': 'aprende python', 'ratings': '4'}
payload = dict(zip(claves, valores))
print(payload)
```
